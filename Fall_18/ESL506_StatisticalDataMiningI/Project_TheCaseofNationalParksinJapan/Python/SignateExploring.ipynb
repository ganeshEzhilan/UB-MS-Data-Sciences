```python
import matplotlib.pyplot as plt
train_df = train.copy()
train_df.index = pd.to_datetime(train_df["datetime"])
train_df = train_df.drop("datetime", axis=1)

```

```python
akan = train_df[train_df["epark"] == "Akan Megumi National Park"]
akan = akan.drop("epark", axis=1)
aso = train_df[train_df["epark"] == "Aso Kuju National Park"]
aso = aso.drop("epark", axis=1)
ise = train_df[train_df["epark"] == "Ise Shima National Park"]
ise = ise.drop("epark", axis=1)
kerama = train_df[train_df["epark"] == "Kerama islands national park"]
kerama = kerama.drop("epark", axis=1)
kirishima = train_df[train_df["epark"] == "Kirishima Jinjiang Bay National Park"]
kirishima = kirishima.drop("epark", axis=1)
nikko = train_df[train_df["epark"] == "Nikko National Park"]
nikko = nikko.drop("epark", axis=1)
oyama = train_df[train_df["epark"] == "Oyama Oki National Park"]
oyama = oyama.drop("epark", axis=1)
towada = train_df[train_df["epark"] == "Towada Hachimantai National Park"]
towada = towada.drop("epark", axis=1)
```

```python
#%matplotlib inline
import matplotlib.pyplot as plt

```

```python


fig = plt.figure(figsize=[16,8])
#plt.plot(akun.resample('W').sum())
plt.plot(aso.resample('W').sum())
#plt.plot(ise.resample('W').sum())
plt.plot(kerama.resample('W').sum())
#plt.plot(kirishima.resample('W').sum())
#plt.plot(nikko.resample('W').sum())
plt.plot(oyama.resample('W').sum())
#plt.plot(towada.resample('W').sum())
#fig.legend(["akun", "aso", "ise", "kerama","kirishima","nikko","oyama","towada"]);
fig.legend(["aso", "kerama","oyama"]);
```

```python
fig = plt.figure(figsize=[16,8])
plt.plot(akan.resample('W').sum())
#plt.plot(aso.resample('W').sum())
plt.plot(ise.resample('W').sum())
#plt.plot(kerama.resample('W').sum())
plt.plot(kirishima.resample('W').sum())
plt.plot(nikko.resample('W').sum())
#plt.plot(oyama.resample('W').sum())
plt.plot(towada.resample('W').sum())
fig.legend(["akan", "ise","kirishima","nikko","towada"]);
#fig.legend(["aso", "kerama","oyama"]);
```

```python
fig = plt.figure(figsize=[16,8])
plt.plot(akan.resample('M').mean())
#plt.plot(aso.resample('W').sum())
plt.plot(ise.resample('M').mean())
#plt.plot(kerama.resample('W').sum())
plt.plot(kirishima.resample('M').mean())
plt.plot(nikko.resample('M').mean())
#plt.plot(oyama.resample('W').sum())
plt.plot(towada.resample('M').mean())
fig.legend(["akan", "ise","kirishima","nikko","towada"]);
#fig.legend(["aso", "kerama","oyama"]);
```

```python
fig = plt.figure(figsize=[16,8])
#plt.plot(akun.resample('W').sum())
plt.plot(aso.resample('M').sum())
#plt.plot(ise.resample('W').sum())
plt.plot(kerama.resample('M').sum())
#plt.plot(kirishima.resample('W').sum())
#plt.plot(nikko.resample('W').sum())
plt.plot(oyama.resample('M').sum())
#plt.plot(towada.resample('W').sum())
#fig.legend(["akun", "aso", "ise", "kerama","kirishima","nikko","oyama","towada"]);
fig.legend(["aso", "kerama","oyama"]);
```

```python
fig = plt.figure(figsize=[16,8])
by_weekday = akan.copy()
by_weekday = by_weekday.groupby(by_weekday.index.dayofweek)
plt.plot(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'],by_weekday.mean())
by_weekday = ise.copy()
by_weekday = by_weekday.groupby(by_weekday.index.dayofweek)
plt.plot(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'],by_weekday.mean())
by_weekday = kirishima.copy()
by_weekday = by_weekday.groupby(by_weekday.index.dayofweek)
plt.plot(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'],by_weekday.mean())
by_weekday = nikko.copy()
by_weekday = by_weekday.groupby(by_weekday.index.dayofweek)
plt.plot(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'],by_weekday.mean())
by_weekday = towada.copy()
by_weekday = by_weekday.groupby(by_weekday.index.dayofweek)
plt.plot(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'],by_weekday.mean())
fig.legend(["akun", "ise","kirishima","nikko","towada"]);
```

```python
fig = plt.figure(figsize=[16,8])
by_weekday = aso.copy()
by_weekday = by_weekday.groupby(by_weekday.index.dayofweek)
plt.plot(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'],by_weekday.mean())
by_weekday = kerama.copy()
by_weekday = by_weekday.groupby(by_weekday.index.dayofweek)
plt.plot(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'],by_weekday.mean())
by_weekday = oyama.copy()
by_weekday = by_weekday.groupby(by_weekday.index.dayofweek)
plt.plot(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'],by_weekday.mean())
fig.legend(["aso", "kerama","oyama"]);
plt.title("Weekly Sampled Data")
```
